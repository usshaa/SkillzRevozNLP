5. Create a BeautifulSoup object to parse the HTML content of the page. `html_content` stores the raw HTML text, and `soup` is the BeautifulSoup object.
6. Use `soup.prettify()` to make the HTML content more readable, although this step is optional.

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Main_Page"
# Add a User-Agent header to mimic a web browser and avoid being blocked
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

response = requests.get(url, headers=headers)
html_content = response.text

soup = BeautifulSoup(html_content, "html.parser")

# Display a snippet of the prettified HTML to confirm successful parsing
display(soup.prettify()[:1000])
```

7. Extract Featured Articles:
   - Use `soup.find_all("div", class_="mp-featured-article")` to locate all elements with the specified class that represent featured articles.
   - Iterate through the found elements, find the article title using `.find("h2").text`, and print each title.

8. Extract "Did You Know" items:
   - Use `soup.find("div", id="mp-dyk")` to locate the section containing "Did You Know" items.
   - Iterate through the list items within this section using `.find_all("li")`, and print each item's text.


```python
# 7. Extract Featured Articles:
# The featured article is typically found within a section with id='mp-tfa' (Today's featured article).
featured_article_section = soup.find("div", id="mp-tfa")

if featured_article_section:
    print("Featured Article:")
    # Find all links within the featured article section
    all_links = featured_article_section.find_all("a")
    found_article = False
    for link in all_links:
        href = link.get('href', '')
        # Check if it's an article link (starts with /wiki/ and is not a special page, disambiguation, or internal link)
        if href.startswith('/wiki/') and ":" not in href and not href.endswith('(disambiguation)') and 'Main_Page' not in href:
            # Ensure the link has some text content and is not just an empty anchor
            if link.text.strip():
                print(f"- {link.text.strip()}")
                found_article = True
                break # Assuming the first relevant link is the main featured article
    if not found_article:
        print("  Could not find a prominent article link within the 'Today's featured article' section.")
else:
    print("Could not find the 'Today's featured article' section (id='mp-tfa').")

print("\n") # Add a newline for separation

# 8. Extract "Did You Know" items:
# The "Did You Know" section is typically found within a div with id='mp-dyk'.
# The items are usually presented as list items (li).
did_you_know_section = soup.find("div", id="mp-dyk")

if did_you_know_section:
    print("Did You Know items:")
    for item in did_you_know_section.find_all("li"):
        print(f"- {item.text.strip()}")
else:
    print("Could not find the 'Did You Know' section (id='mp-dyk').")
```

9. Error Handling:
   - Before sending the request, a try-except block is used to catch any potential HTTP errors using `requests.exceptions.HTTPError`.

10. Exception Handling:
   - Inside the try-except block, there's another try-except block to handle possible attribute errors that may occur when parsing the HTML.


```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Main_Page"
# Add a User-Agent header to mimic a web browser and avoid being blocked
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
except requests.exceptions.HTTPError as e:
    print(e)
    exit()

html_content = response.text

try:
    soup = BeautifulSoup(html_content, "html.parser")

    # 7. Extract Featured Articles:
    # The featured article is typically found within a section with id='mp-tfa' (Today's featured article).
    featured_article_section = soup.find("div", id="mp-tfa")

    if featured_article_section:
        print("Featured Article:")
        all_links = featured_article_section.find_all("a")
        found_article = False
        for link in all_links:
            href = link.get('href', '')
            if href.startswith('/wiki/') and ":" not in href and not href.endswith('(disambiguation)') and 'Main_Page' not in href:
                if link.text.strip():
                    print(f"- {link.text.strip()}")
                    found_article = True
                    break
        if not found_article:
            print("  Could not find a prominent article link within the 'Today's featured article' section.")
    else:
        print("Could not find the 'Today's featured article' section (id='mp-tfa').")

    print("\n")

    # 8. Extract "Did You Know" items:
    # The "Did You Know" section is typically found within a div with id='mp-dyk'.
    did_you_know_section = soup.find("div", id="mp-dyk")

    if did_you_know_section:
        print("Did You Know items:")
        for item in did_you_know_section.find_all("li"):
            print(f"- {item.text.strip()}")
    else:
        print("Could not find the 'Did You Know' section (id='mp-dyk').")

except AttributeError as error:
    print(error)
    exit()
```

11. Saving Data:
   - The extracted data, including featured article titles and "Did You Know" items, is stored in a Python dictionary called `data`.
   - This data is then saved as a JSON file named "data.json" using the `json.dump(data, outfile)` method.
   
12. The script includes proper error handling and exit statements to gracefully handle HTTP errors and attribute errors during the web scraping process.

```python
import requests
from bs4 import BeautifulSoup
import json

url = "https://en.wikipedia.org/wiki/Main_Page"
# Add a User-Agent header to mimic a web browser and avoid being blocked
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
except requests.exceptions.HTTPError as error:
    print(error)
    exit()

html_content = response.text

try:
    soup = BeautifulSoup(html_content, "html.parser")

    # Extract Featured Articles:
    featured_article_section = soup.find("div", id="mp-tfa")
    article_titles = []
    if featured_article_section:
        all_links = featured_article_section.find_all("a")
        for link in all_links:
            href = link.get('href', '')
            if href.startswith('/wiki/') and ":" not in href and not href.endswith('(disambiguation)') and 'Main_Page' not in href:
                if link.text.strip():
                    article_titles.append(link.text.strip())
                    break # Assuming the first relevant link is the main featured article

    # Extract "Did You Know" items:
    did_you_know_section = soup.find("div", id="mp-dyk")
    items = []
    if did_you_know_section:
        items = [item.text for item in did_you_know_section.find_all("li")]

    data = {"featured_articles": article_titles, "did_you_know": items}

    with open("usha.json", "w") as outfile:
        json.dump(data, outfile)

except AttributeError as error:
    print(error)
    exit()
```
